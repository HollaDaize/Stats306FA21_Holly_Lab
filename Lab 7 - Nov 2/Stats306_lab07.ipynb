### Anchors

By default regular expressions will match any part of a string, but often we just want things at the begining or the end.

* `^` Will match the start of a string
*  `$` Will match the end of a string

```r
names <- c("Captain America", "Black Widow", "Antman", "Spiderman", "Thor", 
           "Hulk", "Iron Man", "Hawkeye") %>% 
        str_sort()

writeLines(names)
str_view(names, "^.") # get first letter of each name
```

```r
str_view(paragraph, "...$")  # get the last three characters on each line
```

```r
#see which lines of the paragraph start with a "t"

str_detect(paragraph, "^t")
```

###Character Classes and Alternatives

We have options which match more than one character but less than "." any character.

```r
myVector <- c("abc ","ABC ", "AbC", "123 ", "A 2 z", "?!.", " ")
```

```r
#\d extracts any digit
str_view_all(myVector, "\\d")
```

```r
#\s extracts any whitespace
str_view_all(myVector, "\\s")
```

```r
#[...] extracts anything in the brackets
str_view_all(myVector, "[aA]")
```

```r
#[^...] extracts everything not the brackets
str_view_all(myVector, "[^aA]")
```

```r
#[[:lower:]] extracts any lower letter
str_view_all(myVector, "[[:lower:]]")
```

```r
#[[:alnum:]] extracts any alpha numeric Character
str_view_all(myVector, "[[:alnum:]]")
```

```r
#[^[:alnum:]] extracts any non alpha numeric Character
str_view_all(myVector, "[^[:alnum:]]")
```

```r
#alternation (.|.) lets you pick between two patterns
str_view_all(myVector, "(a|A)(b|B)(c|C)")

str_view_all(myVector, "[aA][bB][cC]")
```

### Repition

What if we want to find all words that have two o's in a row?

We have tools for controlling repition:

*  `?`: 0 or 1
*  `+`: 1 or more
*  `*`: 0 or more
*  `{n}`: exactly n times
*  `{n,}`: n or more times
*  `{,m}`: m or fewer times
*  `{n,m}`: between n and m. [n,m]



```r
## all words with two o's

# .* <- get any and all letters at the begining
# o{2} <- two o's
# .* <- get any and all letters at the end

str_extract(words, ".*o{2}.*") %>% na.omit()
```

```r
## all words with 3 or more vowels in a row

str_extract(words, ".*[aeiou]{3}.*") %>% na.omit()
```

```r
## Webscrapping
# Don't feel like you have to know what all the functions with html in them do

htmlSourceCode <- read_html("https://www.purewow.com/news/the-bachelor-ranking")
htmlSourceCode
```

```r
## Gets all the subheads
ABC_Bachelors <- htmlSourceCode %>%
    html_nodes('h2') %>% 
    html_text

ABC_Bachelors
```

```r
#Extract Names
ABC_Bachelors <- ABC_Bachelors %>%
  str_extract("[[:alpha:]]+ ([[:alpha:]]')?[[:alpha:]]*( [[:alpha:]]{2}\\.)?$") %>%
  na.omit()
writeLines(ABC_Bachelors)
```

### Grouping and Backreferences

Parenthesis provide or statements: `gr(a|e)y` would match with "gray" or "grey".
Parenthesis also provide grouping. You can number certain groups of your regex by putting them in parenthesis without the "or" symbol.

```r
##find all backelor winners that have two sets of double letters in their name

ABC_Bachelors_Winners %>% str_extract("^.*(.)\\1.*(.)\\2.*$") %>% na.omit
```

```r
#find all words that start and end with the same letter

words %>% str_extract("^(.).*\\1$") %>% na.omit
```

### `str_replace()` and `str_replace_all()`

`str_replace()` and `str_replace_all()` allow you to replace matches with new strings.



```r
str_replace(names, "[A-Z]", "<>")
```

```r
str_replace_all("I like to eat, eat, eat. apples and banana!", "[aeiouI]", "a")
str_replace_all("I like to eat, eat, eat. apples and banana!", "[aeiouI]", "e")
str_replace_all("I like to eat, eat, eat. apples and banana!", "[aeiouI]", "o")
str_replace_all("I like to eat, eat, eat. apples and banana!", "[aeiouI]", "i")
str_replace_all("I like to eat, eat, eat. apples and banana!", "[aeiouI]", "u")
```

When we want to change the order of the string, we want to break it into pieces and replace the pieces in seperate orders.

```r
do_sentences <- tibble(sentences = sentences) %>%
    mutate(lower = str_to_lower(sentences)) %>%
    mutate(have_do = str_detect(lower, "\\bdo\\b[^.?]")) %>% 
    filter(have_do)

#Yoda talk
do_sentences$lower %>% 
    str_replace("^(.*[^ ]?) ?do ([^.?]*).", "\\2 do \\1") %>%
    str_to_sentence %>%
    writeLines
```
