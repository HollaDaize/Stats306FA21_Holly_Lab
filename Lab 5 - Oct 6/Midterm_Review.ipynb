**Aesthetic**- a visual property of the objects in your plot. Aesthetics include things like the size, the shape, or the color of your points. Different asethitics are available in different geoms (use ?geom_bar to see what asthetics are available).

**Scaling vs. Manual assignments.** Arguments inside of the aes function will scale the data, how the asthetics appear is directly based on the data values (i.e. color based on diamond cut, shape based on clarity). If you want to manually assign all the data points to the same asethetic (i.e. all points red, triangles, dashed lines), it must be done outside the aes function.

## `gather` and `spread`

Remember that `gather` transforms datasets from "wide" to "long," collecting different column names into a variable. `spread` does the opposite, turning "long" datasets into "wide" ones by spreading out a variable's values into the column names.

```r
require(tidyverse)
```

```r
grades_wide = tribble(
  ~name,  ~`2015`, ~`2016`, ~`2017`,
'Roger',       83,      89,      93,
  'Jon',       92,      90,      93)
grades_wide
```

```r
grades_long = grades_wide %>% gather(key=year, value=grade, `2015`:`2017`)
grades_long
```

```r
grades_long %>% spread(key=year, value=grade)
```
